```python
import json
from regraph import NXGraph, Rule
from regraph import plot_graph, plot_instance, plot_rule
```

```python
# Define the left-hand side of the rule
pattern = NXGraph()
pattern.add_nodes_from([1, 2, 3])
pattern.add_edges_from([(1, 2), (2, 3)])

# create a rule
rule1 = Rule.from_transform(pattern)

# clone a node
p_clone, rsh_clone = rule1.inject_clone_node(1)
print(rule1.to_commands())

# look up cloned nodes
print(rule1.cloned_nodes())
plot_rule(rule1)
```

```python
# delete a node
rule1.inject_remove_node(rsh_clone)
plot_rule(rule1)
```
